# Traffic Crashes & Speed Cameras Overlay Node Visualization

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file = 'Speeding_Crashes.csv'
G = nx.Graph()
```

### Adding Traffic Crashes

```python
df = pd.read_csv(file)
added_nodes = 0
for i in range(0, len(df)):
    if np.isnan(df['LONGITUDE'][i]) or (df['LONGITUDE'][i] == 0):
        continue
    else:
        G.add_node(added_nodes, 
               pos=(df['LONGITUDE'][i], df['LATITUDE'][i]),
               lat=df['LATITUDE'][i], 
               lon=df['LONGITUDE'][i],
               color='grey',
               size=2)
        
        added_nodes+=1
```

```python
file2 = 'speed_cameras.csv'
```

```python
df = pd.read_csv(file2)
added_nodes = 0
for i in range(0, len(df)):
    if np.isnan(df['LONGITUDE'][i]) or (df['LONGITUDE'][i] == 0):
        continue
    else:
        G.add_node(added_nodes, 
               pos=(df['LONGITUDE'][i], df['LATITUDE'][i]),
               lat=df['LATITUDE'][i], 
               lon=df['LONGITUDE'][i],
               color='red',
               size=5)
        added_nodes+=1
```

### Plotting Graph

```python
pos=nx.get_node_attributes(G,'pos')
color=[c for c in nx.get_node_attributes(G,'color').values()]
size=[s for s in nx.get_node_attributes(G,'size').values()]

# Draw and save graph
nx.draw(G, pos, node_size=size, node_color=color, alpha=0.3)
```
