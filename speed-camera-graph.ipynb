# Traffic Crashes & Speed Cameras Overlay Node Visualization

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file = 'Data/Speeding_Crashes.csv'
file2 = 'Data/speed_cameras.csv'
```

### Adding Traffic Crashes

```python
# Crashes
G = nx.Graph()
df = pd.read_csv(file)
added_nodes = 0
for i in range(0, len(df)):
    if np.isnan(df['LONGITUDE'][i]) or (df['LONGITUDE'][i] == 0):
        continue
    else:
        G.add_node(added_nodes, 
               pos=(df['LONGITUDE'][i], df['LATITUDE'][i]),
               lat=df['LATITUDE'][i], 
               lon=df['LONGITUDE'][i],
               color='grey',
               size=2,
               alpha=0.5)
        
        added_nodes+=1
```

```python
# Speed Cameras
H=nx.Graph()
df = pd.read_csv(file2)
added_nodes = 0
for i in range(0, len(df)):
    if np.isnan(df['LONGITUDE'][i]) or (df['LONGITUDE'][i] == 0):
        continue
    else:
        H.add_node(added_nodes, 
               pos=(df['LONGITUDE'][i], df['LATITUDE'][i]),
               lat=df['LATITUDE'][i], 
               lon=df['LONGITUDE'][i],
               color='red',
               size=15,
               alpha=1)
        added_nodes+=1
```

### Plotting Graph

```python
G_pos=nx.get_node_attributes(G,'pos')
H_pos=nx.get_node_attributes(H,'pos')

# Draw and save graph
nx.draw(G, pos=G_pos, node_size=2, node_color='grey', alpha=0.5)
nx.draw_networkx_nodes(H, pos=H_pos,  node_size=3, node_color='red')

plt.savefig("Graphs/Crash Distribution with Speed Cameras.png", dpi=1000)
```
